# Choleshy decomposition



```python
import numpy as np
```

```python
def pos_def_check(A):
	'''
	Checks if a matrix is positive definate
	'''
	pd_state = True
	n = len(A)
	for i in range(1, n):
		if (A[i, i - 1] ** 2) >= A[i, i]:
			raise Exception ("Marix is not positive definate")
			pd_state =  False
	return pd_state
```

```python
def cholesky(A):
	'''
	Function to perform a Cholesky decompostion
	'''
	pd = pos_def_check(A)
	if pd == True:
		n = len(A)
		L = np.zeros_like(A, dtype=np.float64)
		for i in range(0, n):
			for j in range(0, i + 1):
				sum_part = 0
				if i == j:
					for k in range(0, j):
						sum_part += L[j, k] ** 2
					L[i, j] = np.sqrt(A[j, j] - sum_part)
				else:
					for k in range(0, j):
						sum_part += L[i, k] * L[j, k]
					L[i, j] = (A[i, j] - sum_part) / L[j, j]
	else:
		L = f"Matrix is not positive definate"
	return L
```

```python
A = np.array([[np.random.randint(1, 100), 2, 6],
            [2, np.random.randint(1, 100), 5],
            [6, 5, np.random.randint(1, 100)]])
```

```python
print(f"A: {A}")
L = cholesky(A)
print(f"L: {L}")
check = A - np.dot(L, L.T)
print("Check: ")
print(check)
```
